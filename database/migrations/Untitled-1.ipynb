```python
from ortools.linear_solver import pywraplp

# Create solver
solver = pywraplp.Solver.CreateSolver('GLOP')

# Decision variables
x1 = solver.NumVar(0, solver.infinity(), 'x1')
x2 = solver.NumVar(0, solver.infinity(), 'x2')
x3 = solver.NumVar(0, solver.infinity(), 'x3')

# Objective function
solver.Maximize(1000*x1 + 600*x2 + 300*x3)

# Constraints
solver.Add(5000*x1 + 2000*x2 + 1000*x3 <= 20000)
solver.Add(x1 + x2 + x3 <= 10)

# Solve
status = solver.Solve()

# Print result
if status == pywraplp.Solver.OPTIMAL:
    print("Optimal Solution Found")
    print("x1 =", x1.solution_value())
    print("x2 =", x2.solution_value())
    print("x3 =", x3.solution_value())
    print("Maximum Exposure =", solver.Objective().Value())
else:
    print("No optimal solution found.")
```

```python
from ortools.linear_solver import pywraplp

solver = pywraplp.Solver.CreateSolver('GLOP')

# Decision variables
x1 = solver.NumVar(0, solver.infinity(), 'x1')
x2 = solver.NumVar(0, solver.infinity(), 'x2')
x3 = solver.NumVar(0, solver.infinity(), 'x3')
x4 = solver.NumVar(0, solver.infinity(), 'x4')

# Objective Function
solver.Maximize(40*x1 + 35*x2 + 30*x3 + 25*x4)

# Constraints
solver.Add(4*x1 + 5*x2 + 3*x3 + 2*x4 <= 200)
solver.Add(3*x1 + 2*x2 + x3 + 2*x4 <= 60)
solver.Add(x1 + x2 + x3 + x4 <= 30)

status = solver.Solve()

if status == pywraplp.Solver.OPTIMAL:
    print("Optimal Solution Found")
    print("Mathematics =", x1.solution_value())
    print("Computer Science =", x2.solution_value())
    print("Business =", x3.solution_value())
    print("Statistics =", x4.solution_value())
    print("Maximum Students =", solver.Objective().Value())
```
